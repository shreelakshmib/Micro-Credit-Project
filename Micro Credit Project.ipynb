```python
#Use z-score to find outliers :
import numpy as np
from scipy.stats import zscore
#data.info()
z=np.abs(zscore(data))
print(np.where(z>6))
```

```python
a = pd.DataFrame(np.where(z>6))
a.shape
```

```python
#remove the outliers.
data_new = data[(z<6).all(axis=1)]
print(data.shape,"\t",data_new.shape)
```

```python
#Feature Selection

X = data_new.drop('label',axis=1)
Y = data_new.iloc[:,-1:]
```

```python
#Since dataset is imbalanced, we will use oversampling. 
# import library
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)

# fit predictor and target variable
x_ros, y_ros = ros.fit_resample(X, Y)

print('Original dataset shape', Y.shape)
print('Resample dataset shape', y_ros.shape)
```

```python
#Check for skewness - Skewness value should not be greater than 0.5
x_ros.skew()
```

```python
#remove skewness using PowerTransformer
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer('yeo-johnson')
X =pd.DataFrame(pt.fit_transform(x_ros),columns = x_ros.columns.values)
```

```python
#Let us check the skewness again.
X.skew()
```

```python
#convert target variable column to a dataframe
Y = pd.DataFrame(y_ros)
```

```python
#Scale the data using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_rescaled = pd.DataFrame(scaler.fit_transform(X),columns = x_ros.columns.values)
```

```python
X_rescaled.shape
```

```python
#Since there are many columns, we will reduce it using PCA.
from sklearn.decomposition import PCA

#to calculate, how many n_components, we should use, let us create a scree plot
pca = PCA().fit(X_rescaled)

plt.rcParams["figure.figsize"] = (15,6)

fig,ax = plt.subplots()
xi = np.arange(1,25,step=1)
y=np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi,y,marker='o',linestyle='--',color='b')

plt.xlabel("Number of Components")
plt.ylabel("Cumulative Proportion Of Variance Explained")

plt.xticks(np.arange(0,24,step=1)) 
plt.title("The number of components required to explain variance")

plt.axhline(y=0.95,color='r',linestyle='-')
#plt.axvline(x=10,linestyle='--',color='green')
plt.text(0.5,0.85,'95% cut-off threshold',color='red',fontsize=16)

ax.grid(axis='x')
plt.show()
```

    From the graph we can see that 8 components lead to 95% threshold. Hence we will use 8 as n_components

```python
#Use PCA for dimensionality reduction :
pca = PCA(n_components=8)
pca.fit(X_rescaled)
X = pd.DataFrame(pca.transform(X_rescaled))
```

```python
X.shape
```

# Model Prediction and Validation

```python
#import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
#a function to calculate random state seed for each models
def calBestRandomStateOf(model):
    max_score=0
    for i in range(40,100):
        x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=.20,random_state=i)
        model.fit(x_train,y_train)
        pred = model.predict(x_test)
        score = accuracy_score(y_test,pred)
        if score>max_score:
            max_score = score
            final_state = i

    return final_state
```

```python
model = [LogisticRegression(),GaussianNB(),DecisionTreeClassifier(),KNeighborsClassifier()]

for i in range(len(model)):
    state = calBestRandomStateOf(model[i])
    x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=.20,random_state=state)
    model[i].fit(x_train,y_train)
    score = model[i].score(x_train,y_train)
    y_pred = model[i].predict(x_test)
    accuracy = accuracy_score(y_test,y_pred)
    classificationReport = classification_report(y_test,y_pred)
    confusionMatrix = confusion_matrix(y_test,y_pred)
    
    print("*************************************************************************************")
    print("Score of ",model[i]," is : ", score)
    print("Accuracy : ",accuracy*100,"% ")
    print("Classification Report : \n")
    print(classificationReport)
    print("Confusion Matrix :\n ")
    print(confusionMatrix)
    print("*************************************************************************************")
```

    We can see that Decision Tree Classifier is performing better than the rest of the models with an accuracy of 94.67%.

```python
#Cross Validation with accuracy :
from sklearn.model_selection import cross_val_score

for i in range(len(model)):
    cv_score = cross_val_score(model[i],X,Y,cv=4,scoring='accuracy')
    print("*************************************************************************************")
    print("Score for ",model[i]," : ")
    print("Score : ", cv_score)
    print("Mean : ", cv_score.mean())
    print("Standard Deviation : ", cv_score.std())
    print("*************************************************************************************")
    print("")
```

```python
#Cross Validation with f1 - score:
from sklearn.model_selection import cross_val_score

for i in range(len(model)):
    cv_score = cross_val_score(model[i],X,Y,cv=4,scoring='f1')
    print("*************************************************************************************")
    print("Score for ",model[i]," : ")
    print("Score : ", cv_score)
    print("Mean : ", cv_score.mean())
    print("Standard Deviation : ", cv_score.std())
    print("*************************************************************************************")
    print("")
```

```python
#GridSearchCV  for Decision Tree Classifier
from sklearn.model_selection import GridSearchCV

param_dict = {"max_depth": [3, None],
              "max_features": [1,2,3,4,5,6,7,8,9],
              "min_samples_leaf": [1,2,3,4,5]
              }

gridsearch = GridSearchCV(DecisionTreeClassifier(),param_dict,n_jobs=-1,pre_dispatch=2)
gridsearch.fit(X,Y)
gridsearch.best_params_
```

```python
#Using the best parameters, build a dtc model:

dtc = DecisionTreeClassifier(max_depth=None,max_features=1,min_samples_leaf=1)
state = calBestRandomStateOf(dtc)
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=.20,random_state = state )

dtc.fit(x_train,y_train)
score = dtc.score(x_train,y_train)
y_pred = dtc.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)

print("Score for DTC : ",score)
print("Accuracy Score for DTC : ",accuracy)
print("Classification Report : \n")
print(classification_report(y_test,y_pred))
print("Confusion Matrix :\n ")
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score

cv_score = cross_val_score(dtc,X,Y,cv=5,scoring='accuracy')
print("*************************************************************************************")
print("Score for ",dtc," : ")
print("Score : ", cv_score)
print("Mean : ", cv_score.mean())
print("Standard Deviation : ", cv_score.std())
print("*************************************************************************************")
print("")
```

```python
#AuC-ROC Curve :
from sklearn.metrics import roc_curve,roc_auc_score
import matplotlib.pyplot as plt

y_pred_prob = dtc.predict_proba(x_test)[:,1]
fpr,tpr,threshold=roc_curve(y_test,y_pred_prob)
auc_score = roc_auc_score(y_test,dtc.predict(x_test))
plt.figure(figsize=(8,5))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,Label='Area under the curve = %0.3f'%auc_score)
plt.legend(loc = "lower right")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("Logistic Regression")
plt.show()
```

    So we have AUC score of about 95.4 which represents our model is performing extremely good.

```python
#Adaboost Regressor and RandomforestRegressor
from sklearn.ensemble import AdaBoostClassifier as ABC
from sklearn.ensemble import GradientBoostingClassifier as GBC
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.ensemble import BaggingClassifier as BC
from sklearn.ensemble import ExtraTreesClassifier as ETC


ada = ABC(n_estimators=20,random_state=41)
gradient = GBC(n_estimators=20,random_state=162)
rfc = RFC(n_estimators=20,random_state=115)
bc = BC(n_estimators=20,random_state=42)
etc = ETC(n_estimators=20,random_state=86)

boosting_model = [ada,gradient,rfc,bc,etc]

for i in range(len(boosting_model)):
    boost = boosting_model[i]
    boost.fit(x_train,y_train)
    pred = boost.predict(x_test)
    accuracyScore = accuracy_score(y_test,pred)
    print("-----------------------------------------------------------")
    print(boost)
    print("-----------------------------------------------------------")
    print("Accuracy Score : ", accuracyScore)
    print("\n")
```

```python
#Boosted model - Extra Tree Classifier performed the best and hence we will build our final model with it.
etc = ETC(n_estimators=20,random_state=162)
etc.fit(x_train,y_train)
y_pred= etc.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy Score for Extra Tree Classifier : ",round(accuracy*100,2),"%")
print("Classification Report : \n")
print(classification_report(y_test,y_pred))
print("Confusion Matrix :\n ")
print(confusion_matrix(y_test,y_pred))
```
